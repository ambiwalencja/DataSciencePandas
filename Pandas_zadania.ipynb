```python
# 1. Załaduj zbiór danych korzystajac z biblioteki Pandas. 
import pandas as pd
df = pd.read_csv('dataset_pandas_pd.csv',
                 delimiter=',',
                 names=['imie', 'nazwisko', 'rok_ur', 'miesiac_ur', 'dzien_ur', 'miejsce_ur', 'miasto', 'departament',
                        'rok_pracy', 'miesiac_pracy', 'dzien_pracy', 'doswiadczenie'])
```

```python
# 2. Osób o jakim nazwisku jest najwięcej ?
print(df.groupby('nazwisko').nazwisko.count().sort_values()[[-1]])
```

```python
# 3. Po kolumnie z nazwiskiem dodaj kolumnę reprezentującą płeć danej osoby. Można wykorzystać fakt, że imiona
# kobiet kończą się literka 'a'.
sex_values = ['K' if name[-1]=='a' else 'M' for name in list(df['imie'])]
df.insert(2, 'plec', sex_values)
df.head()
```

```python
# 4. Posortuj zbiór danych alfabetycznie po nazwiskach, a w ramach jednego nazwiska po roku urodzenia.
df.sort_values(by=['nazwisko', 'rok_ur'])
```

```python
# 5. Do zbioru danych należy dodać kolumnę, która będzie mówiła o poziomie doświadczenia danej osoby, w sposób
# słowny, przyjmując:

# Junior (do 3 lat doświadczenia)
# Mid (4 do 8 lat)
# Senior (9 do 12 lat)
# Ekspert (od 13 lat)

# 1)
experience_levels = []
for experience in list(df['doswiadczenie']):
    if experience <= 3:
        experience_levels.append('Junior')
    elif experience in range(4,9):
        experience_levels.append('Mid')
    elif experience in range(9,13):
        experience_levels.append('Senior')
    else:
        experience_levels.append('Ekspert')
df['poziom_dosw'] = experience_levels
```

```python
# 5
# sposób 2) przy użyciu loc()
df['poziom_dosw'] = 'Junior'
df.loc[df['doswiadczenie']>3,'poziom_dosw'] = 'Mid'
df.loc[df['doswiadczenie']>8,'poziom_dosw'] = 'Senior'
df.loc[df['doswiadczenie']>12,'poziom_dosw'] = 'Ekspert'

df.head()
```

```python
# 5.
# sposób 3) przy użyciu apply()
def experience_levels(x):
    if x <= 3:
        return 'Junior'
    elif x in range(4,9):
        return 'Mid'
    elif x in range(9,13):
        return'Senior'
    else:
        return 'Ekspert'

df['poziom_dosw'] = df['doswiadczenie'].apply(experience_levels)
```

```python
# 5. 
# 4) sposób oparty na np.select()
import numpy as np
experience_conditions = [
    (df['doswiadczenie'] < 4),
    ((df['doswiadczenie'] >= 4) & (df['doswiadczenie'] < 9)),
    ((df['doswiadczenie'] >= 9) & (df['doswiadczenie'] < 13)),
    (df['doswiadczenie']> 13)
]
experience_levels = ['Junior', 'Mid', 'Senior', 'Ekspert']
df['poziom_dosw'] = np.select(experience_conditions, experience_levels)
```

```python
# 6. Kolumna reprezentująca rok dołączenia do firmy ma puste wartości. Uzupełnij brakujące wartości używając
# średniej dla osób o takim samym poziomie doświadczenia (kolumna z Zadania 3), w skali całej firmy.

junior_mean = df[df['poziom_dosw'] == 'Junior'].rok_pracy.mean()
mid_mean = round(df[df['poziom_dosw'] == 'Mid'].rok_pracy.mean())
senior_mean = round(df[df['poziom_dosw'] == 'Senior'].rok_pracy.mean())
expert_mean = round(df[df['poziom_dosw'] == 'Ekspert'].rok_pracy.mean())

def _mean(x):
  if x == 'Ekspert':
    return expert_mean 
  if x == 'Senior':
    return senior_mean 
  if x == 'Mid':
    return mid_mean 
  if x == 'Junior':
    return junior_mean

df['rok_pracy'].fillna(df['poziom_dosw'].apply(_mean), inplace=True)

print(mid_mean)
print(senior_mean)
print(expert_mean)
print('\n')
df.head(20)
```

```python
# 6.
# sposób 2, dynamiczny - groupby + dict + map
levels_dict = df.groupby('poziom_dosw')['rok_pracy'].mean().round().to_dict()
df['rok_pracy'] = df['rok_pracy'].fillna(df['poziom_dosw'].map(levels_dict))

df.head(20)
```

```python
# 7. Który oddział firmy ma najwięcej Data Scientistów ?
df[df['departament'] == 'Data Science'].groupby('miasto').departament.count().sort_values()[[-1]]
```

```python
# 8. Szef chciałby zmienić nazewnictwo departamentów w całej firmie. Napisz kod, który zamieni wartości w
# kolumnie reprezentującej departament, nastepująco:

# Software Development -> Software Development Team
# HR -> Human Resources
# Data Science -> Data Analytics

df.loc[df["departament"] == "Software Development", "departament"] = 'Software Development Team'
df.loc[df["departament"] == "HR", "departament"] = 'Human Resources'
df.loc[df["departament"] == "Data Science", "departament"] = 'Data Analytics'

df.head()
```

```python
# 9. Które biuro firmy ma najwięcej osób na Seniorskim poziomie doświadczenia ?
df[df['poziom_dosw'] == 'Senior'].groupby('miasto').departament.count().sort_values()[[-1]]
```

```python
# 10. Które biuro ma najwięcej kobiet na Juniorskim poziomie doświadczenia ?
df[(df['poziom_dosw'] == 'Junior') & (df['plec'] == 'K')].groupby('miasto').departament.count().sort_values()#[[-1]]
```

```python
# 11. Ile jest kobiet w krakowskim biurze, na Mid poziomie doświadczenia, urodzonych przed 1975 rokiem ?
df[(df['miasto'] == 'Krakow') & (df['poziom_dosw'] == 'Mid') & (df['plec'] == 'K') & (df['rok_ur'] < 1975)]['imie'].count()
```

```python
# 12. Ile osób, które posiadają obecnie stanowisko seniorskie, dołączyło do firmy w przeciągu ostatnich 10 lat?
from datetime import date
df[(df['poziom_dosw'] == 'Senior') & (date.today().year - df['rok_pracy'] < 10)]['imie'].count()
```
